```python
import numpy as np
import pandas as pd
import torch
import matplotlib.pyplot as plt
from matplotlib import gridspec
from sentence_transformers import SentenceTransformer
from transformers import pipeline
from river import drift, naive_bayes, ensemble, metrics

sbert_model = SentenceTransformer('paraphrase-MiniLM-L6-v2')


df = pd.read_csv('small_sample.csv')


sentiment_model = pipeline("sentiment-analysis", model="cardiffnlp/twitter-roberta-base-sentiment")
df['sentiment'] = df['cleaned_text'].apply(lambda text: sentiment_model(text)[0]['label'] if isinstance(text, str) else 'neutral')


label_map = {'negative': 0, 'neutral': 1, 'positive': 2}
df['sentiment_label'] = df['sentiment'].map(label_map)


def class_swap(label):
    swap_dict = {"positive": "negative", "negative": "positive"}
    return swap_dict.get(label, label)

df['modified_sentiment'] = df['sentiment'].apply(class_swap)

def class_shift(index, label):
    return "positive" if index % 5000 < 2500 else "negative"
df['class_shifted'] = df.apply(lambda row: class_shift(row.name, row['sentiment']), axis=1)


df['embeddings'] = df['cleaned_text'].apply(lambda text: sbert_model.encode(text) if isinstance(text, str) else np.zeros(384))

df_time_sliced = df.iloc[int(len(df) * 0.1):].reset_index(drop=True)

drift_detectors = {"modified_sentiment": drift.ADWIN(), "class_shifted": drift.ADWIN()}
drifts = {"modified_sentiment": [], "class_shifted": []}

clf = ensemble.AdaptiveRandomForestClassifier()
metric = metrics.MacroF1()
macro_f1_scores = []
iterations = []

for i, row in df.iterrows():
    x = {'embedding_mean': np.mean(row['embeddings'])}
    y = row['sentiment_label']
    y_pred = clf.predict_one(x)
    clf.learn_one(x, y)

    if y_pred is not None:
        metric.update(y, y_pred)
        macro_f1_scores.append(metric.get())
        iterations.append(i)

    for col in ["modified_sentiment", "class_shifted"]:
        encoded_value = label_map.get(row[col], 1)
        drift_detectors[col].update(encoded_value)
        if drift_detectors[col].change_detected:
            print(f'Change detected in {col} at index {i}')
            drifts[col].append(i)
            drift_detectors[col].reset()


fig, axes = plt.subplots(3, 1, figsize=(10, 8))


axes[0].grid()
axes[0].plot(range(len(df)), df['sentiment_label'], label='Original Sentiment')
for drift_detected in drifts["modified_sentiment"]:
    axes[0].axvline(drift_detected, color='red')
axes[0].set_title("Drift Detection in Modified Sentiment")


axes[1].grid()
axes[1].plot(range(len(df)), df['sentiment_label'], label='Original Sentiment')
for drift_detected in drifts["class_shifted"]:
    axes[1].axvline(drift_detected, color='red')
axes[1].set_title("Drift Detection in Class Shifted")


axes[2].grid()
axes[2].plot(iterations, macro_f1_scores, label='Macro-F1 Score', color='blue')
axes[2].set_title("Macro-F1 Score vs Iteration")
axes[2].set_xlabel("Iteration")
axes[2].set_ylabel("Macro-F1 Score")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import nltk
import random
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from sentence_transformers import SentenceTransformer
from river import drift


df = pd.read_csv("small_sample.csv")
df = df.dropna(subset=['cleaned_text'])


def get_antonym(word):
    antonyms = []
    for syn in wordnet.synsets(word, pos=wordnet.ADJ):
        for lemma in syn.lemmas():
            if lemma.antonyms():
                antonyms.append(lemma.antonyms()[0].name())
    return antonyms[0] if antonyms else word

def adjective_swap(sentence):
    words = word_tokenize(sentence)
    pos_tags = pos_tag(words)

    swapped_sentence = []
    for word, tag in pos_tags:
        if tag in ['JJ', 'JJR', 'JJS']:
            swapped_sentence.append(get_antonym(word))
        else:
            swapped_sentence.append(word)

    return " ".join(swapped_sentence)



drift_indices = [
    int(len(df) * 0.2),
    int(len(df) * 0.5),
    int(len(df) * 0.8)
]


df['drift_text'] = df['cleaned_text']

for idx in drift_indices:
    df.loc[idx:idx+550, 'drift_text'] = df.loc[idx:idx+550, 'cleaned_text'].apply(adjective_swap)

print("Drift applied to samples at:",[(idx,idx+550) for idx in drift_indices])


sbert_model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

df['original_embeddings'] = df['cleaned_text'].apply(lambda text: sbert_model.encode(text) if isinstance(text, str) else np.zeros(384))
df['drift_embeddings'] = df['drift_text'].apply(lambda text: sbert_model.encode(text) if isinstance(text, str) else np.zeros(384))


df['original_embeddings_mean'] = df['original_embeddings'].apply(lambda emb: np.mean(emb))
df['drift_embeddings_mean'] = df['drift_embeddings'].apply(lambda emb: np.mean(emb))


print("First few mean embedding values (original):", df['original_embeddings_mean'].head())
print("First few mean embedding values (drifted):", df['drift_embeddings_mean'].head())

drift_detector = drift.ADWIN()
drift_points = []

for i, row in df.iterrows():
    embedding_mean = row['drift_embeddings_mean']
    drift_detector.update(embedding_mean)

    if drift_detector.change_detected:
        print(f"Drift detected at index {i}, Mean embedding: {embedding_mean}")
        drift_points.append(i)
        drift_detector.reset()

print("\nFinal drift points detected:", drift_points)
```
